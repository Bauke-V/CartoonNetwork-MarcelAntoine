```python
import pandas as pd

df_ner['article_number'] = df_ner['article_number']
df_overzicht_expanded = df_ner[['IDN', 'Date', 'Page', 'article_number']].drop_duplicates()

df_overzicht_expanded = df_overzicht_expanded.merge(
    df_overzicht,
    on=['IDN', 'Date', 'Page'],
    how='left'
)

merged_df = df_overzicht_expanded.merge(
    df_ner,
    on=['IDN', 'Date', 'Page', 'article_number'],
    how='left',
    suffixes=('_meta', '_ner')
)

merged_df = merged_df.sort_values(by=['IDN', 'Date', 'Page', 'article_number']).reset_index(drop=True)

merged_df.head()
```

```python
metadata_cols = ['BP-date', 'Title in BP', 'Title', 'Commentaire', 'Categories_test', 'Image', 'Lien KBR']

missing_all_metadata = merged_df[metadata_cols].isnull().all(axis=1)
print(f"Articles missing ALL metadata: {missing_all_metadata.sum()}")

missing_any_metadata = merged_df[metadata_cols].isnull().any(axis=1)
print(f"Articles missing ANY metadata field: {missing_any_metadata.sum()}")

print("\nSample of articles with partial or full missing metadata:")
print(merged_df[missing_any_metadata].head())

print("\nMissing value count per metadata column:")
print(merged_df[metadata_cols].isnull().sum())
```

```python
merged_df['Categories_test'] = merged_df['Categories_test'].replace('Radio+Slache', 'Radio')
merged_df = merged_df[
    merged_df['_ - people'].notna() & 
    (merged_df['_ - people'] != 'None') & 
    (merged_df['_ - people'].str.strip() != '')
]
```

```python
merged_df.to_csv("merged_articles_with_metadata_Radio.csv", index=False)
```

```python
import itertools
import pandas as pd

valid_categories = ['MineSouriante', 'MA-subj', 'Radio']
df_filtered = merged_df[merged_df['Categories_test'].isin(valid_categories)]

df_people = df_filtered[df_filtered['_ - people'].notna()].copy()

df_people['people_list'] = df_people['_ - people'].str.split('|')
df_people['people_list'] = df_people['people_list'].apply(
    lambda lst: [p.split(':')[0] for p in lst if p and p != 'None']
)

edges = []
for _, row in df_people.iterrows():
    people = row['people_list']
    category = row['Categories_test']
    if len(people) > 1:
        for source, target in itertools.combinations(sorted(set(people)), 2):
            edges.append((source, target, category))

edges_df = pd.DataFrame(edges, columns=['Source', 'Target', 'Category'])
edges_df['Weight'] = 1
edges_df = edges_df.groupby(['Source', 'Target', 'Category'], as_index=False).count()

nodes = pd.DataFrame(pd.unique(edges_df[['Source', 'Target']].values.ravel('K')), columns=['Id'])
nodes['Label'] = nodes['Id']

edges_df.to_csv('edges_marcel_Categories.csv', index=False)
nodes.to_csv('nodes_Marcel_Categories.csv', index=False)
```
